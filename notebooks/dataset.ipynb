```python
import json
from math import ceil, floor

import numpy as np
from PIL import Image
from pathlib import Path
from tqdm import tqdm
```

```python
FULLFOLDER = '../datasets/raw/full/images'
PARTFOLDER = '../datasets/raw/part/images'
img_paths = list(Path(FULLFOLDER).glob('*.jpg')) + list(Path(PARTFOLDER).glob('*.jpg'))
img_paths
```

```python
testpath = str(img_paths[0])
testpath
```

```python
testpath[testpath.rfind('/')+1:]
```

```python
trainval = np.random.choice(['train', 'val'], p=[0.9, 0.1], size=len(img_paths))
trainval
```

```python
for subfolder, img_path in tqdm(zip(trainval, img_paths), total=len(img_paths)):
    img_path = str(img_path)
    img_name = img_path[img_path.rfind('/')+1:]
    img = Image.open(img_path)
    with open(img_path.replace('.jpg', '.json').replace('images', 'labels'), 'r') as f:
        label_dict = json.load(f)
    c = label_dict[0]['annotations'][0]['coordinates']
    rad_h = ceil(c['height'] / 2)
    rad_w = ceil(c['width'] / 2)
    crop_img = img.crop((c['x'] - rad_w, c['y'] - rad_h, c['x'] + rad_w, c['y'] + rad_h))
    crop_img.save(f'../datasets/crops/{subfolder}/{img_name}')
```
